```python
from tastytrade.option import Option, OptionChain
from tastytrade.session import Session
from tastytrade.streamer import DataStreamer, EventType
from tastytrade.account import Account
```

```python
session = Session('graemecracker', ':4s-S9/9L&Q~C]@v')
streamer = await DataStreamer.create(session)
```

```python
quotes = await streamer.stream(EventType.QUOTE, ['SPY'])
quotes
```

```python
chain = await OptionChain.fetch(session, 'SPY')
```

```python
exps = chain.get_expirations()
chain.get_chain_at(exps[-3])
```

```python
accs = await Account.get_accounts(session)
accs
```

```python
bal = await accs[0].get_balance(session)
bal2 = await accs[1].get_balance(session)
bal, bal2
```
